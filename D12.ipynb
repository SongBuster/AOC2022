```python
class TreeNode:
    def __init__(self, data, parent):
        self.data = data
        self.children = []
        self.parent = parent
        self.is_end = False

    def add_children(self, leaf):        
        self.children.append(leaf)
        
    def __str__(self):
        return f"{self.data}{self.is_end}"    

class Tree:
    def __init__(self, data):
        self.root = TreeNode(data, None)


def print_leaf(leaf, prof):
    chars = '-' * prof
    print (f"{chars}{leaf.data}")
    new_prof = prof + 1
    for new_leaf in leaf.children:
        print_leaf(new_leaf,new_prof)
    

def print_tree(tree):
    print (tree.root.data)
    for leaf in tree.root.children:
        print_leaf(leaf,1)

def print_node(node):
    print (node.data)
    for leaf in node.children:
        print_leaf(leaf,1)
```

```python
def find_start(map):
    for row in range(0,len(map)):
        if 'S' in map[row]:
            return (row, map[row].index('S'))

def get_map_height(map, point):
    return map[point[1]][point[0]]


def comp_points_height(map,point1, point2):
    # returns difference between point2 height minus point1 height
    # positive then point2 is higher than point1
    # negative then point2 is lower than point1
    h1 = get_map_height(map, point1)
    h2 = get_map_height(map, point2)
    #print(f"ccc {h2} - {h1}")
    if h1 == 'S': 
        h1 = 'a'
    if h1 == 'E':
        h1 = 'z'    # change to z!!!!!!
    if h2 == 'S':
        h2 = 'a'
    if h2 == 'E':
        h2 = 'z'    # change to z!!!!!!
    if h1 == 'X':
        return -1
    if h2 == 'X':
        return 999
    
    return ord(h2)-ord(h1)

def go_right(point,map):    
    #print(f"{point[0] + 1 >= len(map[point[1]])}")
    #print(f"{comp_points_height(map,point,(point[0]+1,point[1])) < 0}")
    if point[0] + 1 >= len(map[point[1]]):
        return None
    elif comp_points_height(map,point,(point[0]+1,point[1])) > 1 or comp_points_height(map,point,(point[0]+1,point[1])) < 0:
        return None
    else:
        return (point[0]+1, point[1])

def go_down(point,map):
    if point[1] + 1 >= len(map):
        return None
    elif comp_points_height(map,point,(point[0],point[1]+1)) > 1 or comp_points_height(map,point,(point[0],point[1]+1)) < 0:
        return None
    else:
        return (point[0], point[1]+1)

def go_left(point,map):
    if point[0] - 1 < 0:
        return None
    elif comp_points_height(map,point,(point[0]-1,point[1])) > 1 or comp_points_height(map,point,(point[0]-1,point[1])) < 0:
        return None
    else:
        return (point[0]-1, point[1])

def go_up(point,map):
    if point[1] - 1 < 0:
        return None
    elif comp_points_height(map,point,(point[0],point[1]-1)) > 1 or comp_points_height(map,point,(point[0],point[1]-1)) < 0:
        return None
    else:
        return (point[0], point[1]-1)

def set_point_visited(map, point):
    row = point[1]
    col = point[0]
    map[row][col] = 'X'
    #print_map(map)

def print_map(map):
    for row in range(0, len(map)):        
        print (''.join(map[row]))

def get_possible_dirs(map, point, visited_points):
    poss_dirs = []

    new_point = go_right(point,map)
    if new_point != None and new_point not in(visited_points):        
        poss_dirs.append(('>',new_point))
    
    new_point = go_down(point,map)
    if new_point != None and new_point not in(visited_points):
        poss_dirs.append(('v',new_point))
    
    new_point = go_left(point,map)
    if new_point != None and new_point not in(visited_points):
        poss_dirs.append(('<',new_point))
    
    new_point = go_up(point,map)
    if new_point != None and new_point not in(visited_points):
        poss_dirs.append(('^',new_point))

    return poss_dirs


def find_ways(map, point, node, visited_points):

    this_vp = deepcopy(visited_points)
    this_vp.append(point)

    if get_map_height(map, point) == 'E':
        node.is_end = True
        return node
    else:
        poss_dirs = get_possible_dirs(map, point, this_vp)        
        #print(poss_dirs)
        if (len(poss_dirs)>0):
            for new_dir in poss_dirs:
                new_point = new_dir[1]
                dir = new_dir[0]
                new_child = find_ways(map, new_point, TreeNode(dir,node), this_vp)                
                node.add_children(new_child)                
        else:
            node.is_end = False

        
        return node
    
def count_depth(node, depth):
    
    if node.is_end:
        depths.append(depth)
        return depth
    else:
        depth += 1
        if len(node.children) > 0:
            for child in node.children:
                depth = count_depth(child, depth)
            return depth
        else:
            return 999
            

    return 0
        
```
